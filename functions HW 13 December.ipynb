## Задание 1

```python
books = [
    {'genre': 'поэзия', 'number': '978-5-1000-1234-7', 'title': 'Евгений Онегин', 'author': 'Александр Пушкин'},
    {'genre': 'фэнтези', 'number': '88006', 'title': 'Властелин колец', 'author': 'Джон Р. Р. Толкин'},
    {'genre': 'детектив', 'number': 'D-1122', 'title': 'Безмолвный свидетель', 'author': 'Агата Кристи'}
]

directories = {
    '1': ['978-5-1000-1234-7', '88006'],
    '2': ['D-1122'],
    '3': []
}

def find_book_by_number(number):
    for book in books:
        if book['number'] == number:
            return book
    return None


def find_shelf_by_book(number):
    for shelf, nums in directories.items():
        if number in nums:
            return shelf
    return None


def print_all_books():
    for book in books:
        shelf = find_shelf_by_book(book['number'])
        print(
            f"№: {book['number']}, жанр: {book['genre']}, "
            f"название: {book['title']}, автор: {book['author']}, "
            f"полка хранения: {shelf}"
        )

def book_info():
    number = input("Введите номер книги: ")
    book = find_book_by_number(number)
    if book:
        print(f"Название книги: {book['title']}")
        print(f"Автор: {book['author']}")
    else:
        print("Книга не найдена в базе")

def shelf():
    title = input("Введите название книги: ")
    for book in books:
        if book['title'] == title:
            shelf_num = find_shelf_by_book(book['number'])
            print(f"Книга хранится на полке: {shelf_num}")
            return
    print("Книга не найдена в базе")

def add_shelf():
    shelf = input("Введите номер полки: ")
    if shelf in directories:
        print(f"Такая полка уже существует. Текущий перечень полок: {', '.join(directories.keys())}.")
    else:
        directories.setdefault(shelf, [])
        print(f"Полка добавлена. Текущий перечень полок: {', '.join(directories.keys())}.")
        
def del_shelf():
    shelf = input("Введите номер полки: ")
    if shelf not in directories:
        print(f"Такой полки не существует. Текущий перечень полок: {', '.join(directories.keys())}.")
    elif directories[shelf]:
        print(f"На полке есть книги, удалите их перед удалением полки. Текущий перечень полок: {', '.join(directories.keys())}.")
    else:
        directories.pop(shelf)
        print(f"Полка удалена. Текущий перечень полок: {', '.join(directories.keys())}.")
```

## Задание 2

```python
def add_book():
    number = input("Введите номер книги: ")
    genre = input("Введите жанр книги: ")
    title = input("Введите название книги: ")
    author = input("Введите автора книги: ")
    shelf = input("Введите полку для хранения: ")

    if shelf not in directories:
        print("Такой полки не существует. Добавьте полку командой add_shelf.")
        print_all_books()
        return

    books.append({
        'genre': genre,
        'number': number,
        'title': title,
        'author': author
    })
    directories[shelf].append(number)

    print("Книга добавлена. Текущий список книг:")
    print_all_books()

def del_book():
    number = input("Введите номер книги: ")
    book = find_book_by_number(number)

    if not book:
        print("Книга не найдена в базе.")
        print_all_books()
        return

    books.remove(book)
    shelf = find_shelf_by_book(number)
    directories[shelf].remove(number)

    print("Книга удалена.")
    print_all_books()

def move():
    number = input("Введите номер книги: ")
    shelf = input("Введите номер полки: ")

    if shelf not in directories:
        print(f"Такой полки не существует. Текущий перечень полок: {', '.join(directories.keys())}.")
        return

    book = find_book_by_number(number)
    if not book:
        print("Книга не найдена в базе.")
        print_all_books()
        return

    old_shelf = find_shelf_by_book(number)
    directories[old_shelf].remove(number)
    directories[shelf].append(number)

    print("Книга перемещена.")
    print_all_books()

while True:
    command = input("\nВведите команду: ")

    match command:
        case "book_info":
            book_info()
        case "shelf":
            shelf()
        case "all":
            print_all_books()
        case "add_shelf":
            add_shelf()
        case "del_shelf":
            del_shelf()
        case "add_book":
            add_book()
        case "del_book":
            del_book()
        case "move":
            move()
        case "q":
            print("Выход из программы")
            break
        case _:
            print("Неизвестная команда")
```
